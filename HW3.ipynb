3. Внизу создайте логарифмический график с отношением полученных площадей 3-мя разными методами к действительной (аналитически посчитанной) области в зависимости от количеста разбиений $n$

$\textcolor{yellow}{[2\;балла]}$



![](task_2.png)

```python
import numpy as np
import matplotlib.pyplot as plt

def func(x):
    return np.sin(x)**2

a = 0
b = 2*np.pi

n = 10

dx = (b-a)/n

heights = func(np.linspace(a+dx/2, b-dx/2, n))

areas = heights*dx

total_area = np.sum(areas)

x = np.linspace(a, b, 100)
fig, ax = plt.subplots(figsize=(8, 8))
ax.plot(x, func(x), 'b', linewidth=2)
for i in range(n):
    rect = plt.Rectangle((a+i*dx,0), dx, heights[i], color='g', alpha=0.4)
    ax.add_patch(rect)
ax.set_xlim([a,b])
ax.set_ylim([0,1])
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('Numerical integration of sin^2(x) using rectangular approximation\n(Rectangles = {})'.format(n))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x)

a = 0
b = np.pi

n = 20

h = (b - a) / n

x_left = np.linspace(a, b - h, n)
x_mid = np.linspace(a + h / 2, b - h / 2, n)
x_right = np.linspace(a + h, b, n)

height_left = f(x_left)
height_mid = f(x_mid)
height_right = f(x_right)

area_left = height_left * h
area_mid = height_mid * h
area_right = height_right * h

total_area_left = np.sum(area_left)
total_area_mid = np.sum(area_mid)
total_area_right = np.sum(area_right)

fig, axs = plt.subplots(2, 3, figsize=(10, 8))

axs[0, 0].plot(x_left, height_left, 'r')
axs[0, 0].bar(x_left, height_left, width=h, alpha=0.2)
axs[0, 0].set_title('Left Rectangle Method')

axs[0, 1].plot(x_mid, height_mid, 'g')
axs[0, 1].bar(x_mid, height_mid, width=h, alpha=0.2)
axs[0, 1].set_title('Middle Rectangle Method')

axs[0, 2].plot(x_right, height_right, 'b')
axs[0, 2].bar(x_right, height_right, width=h, alpha=0.2)
axs[0, 2].set_title('Right Rectangle Method')

axs[1, 0].bar(['Left', 'Middle', 'Right'], [total_area_left, total_area_mid, total_area_right], alpha=0.5)
axs[1, 0].set_title('Total Area')

fig.delaxes(axs[1, 1])

fig.tight_layout()

plt.show()
```

# Задание 3 $\textcolor{yellow}{[6\;баллов]}$

"Разбросайте" 100 разноцветных точек по плоскости [таким образом](https://matplotlib.org/stable/gallery/axes_grid1/scatter_hist_locatable_axes.html#sphx-glr-gallery-axes-grid1-scatter-hist-locatable-axes-py), чтобы:
1. По оси $x$ [нормальный](https://numpy.org/doc/stable/reference/random/generated/numpy.random.randn.html) разброс был в 2 раза больше, чем по оси $y$ $\textcolor{yellow}{[2\;балла]}$
2. Были построены гистограммы распределения точек по координатам $\textcolor{yellow}{[2\;балла]}$
3. Можно было [управлять](https://numpy.org/doc/stable/reference/random/generated/numpy.random.choice.html) соотношением цветов. И соответсвующее соотношение цветов (количество точек каждого цвета) показывалось бы в столбчатой диаграмме $\textcolor{yellow}{[2\;балла]}$ 

![](task_3.jpg)

```python
import numpy as np
import matplotlib.pyplot as plt

cov = [[4, 0], [0, 2]]
x, y = np.random.multivariate_normal([0, 0], cov, 100).T

color_ratio = [0.2, 0.3, 0.5]
COLORS = ['red', 'green', 'blue']
num_colors = len(color_ratio)
colors = np.random.choice(num_colors, size=100, p=color_ratio)

color_counts = []
for i in range(num_colors):
    color_counts.append(np.sum(colors==i))

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10,5))
ax1.hist(x, bins=20, color='blue')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Count')
ax2.hist(y, bins=20, color='blue')
ax2.set_xlabel('Y-axis')
ax2.set_ylabel('Count')

fig, ax3 = plt.subplots(figsize=(7,7))
ax3.scatter(x, y, c=np.random.choice(COLORS, size=100), alpha=1)
ax3.set_xlabel('X-axis')
ax3.set_ylabel('Y-axis')

fig, ax4 = plt.subplots()
ax4.bar(range(num_colors), color_counts, color=COLORS)
ax4.set_xticks(range(num_colors))
ax4.set_xlabel('Color')
ax4.set_ylabel('Count')

plt.show()
```
